### Feature distributions by class
To better understand how each predictor behaves across classes, the following charts combine class-conditional histograms with boxplots for every feature.


```python
feature_columns = [col for col in df.columns if col != 'target_class']

for feature in feature_columns:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    sns.histplot(
        data=df,
        x=feature,
        hue='target_class',
        stat='density',
        common_norm=False,
        element='step',
        bins=30,
        ax=axes[0],
    )
    axes[0].set_title(f"Distribution of {feature} by class")
    axes[0].set_xlabel(feature)
    axes[0].set_ylabel("Density")

    sns.boxplot(data=df, x='target_class', y=feature, ax=axes[1])
    axes[1].set_title(f"Boxplot of {feature} by class")
    axes[1].set_xlabel('target_class')
    axes[1].set_ylabel(feature)

    plt.tight_layout()
    plt.show()
```

### IQR-based outlier counts
The table below counts observations beyond 1.5  IQR from the first and third quartiles within each class, offering a simple view of how extreme values are distributed.


```python
outlier_rows = []
for feature in feature_columns:
    for cls, subset in df.groupby('target_class'):
        q1 = subset[feature].quantile(0.25)
        q3 = subset[feature].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        mask = (subset[feature] < lower) | (subset[feature] > upper)
        count = int(mask.sum())
        total = subset.shape[0]
        pct = (count / total) if total else float('nan')
        outlier_rows.append(
            {
                'feature': feature,
                'target_class': cls,
                'outlier_count': count,
                'outlier_pct': pct,
            }
        )

outlier_df = pd.DataFrame(outlier_rows)
outlier_table = outlier_df.pivot(index='feature', columns='target_class', values=['outlier_count', 'outlier_pct'])
outlier_table = outlier_table.sort_index(axis=1, level=1)
outlier_table.columns = [f"class_{cls}_{metric}" for metric, cls in outlier_table.columns]
outlier_table = outlier_table.sort_index()
outlier_table = outlier_table.round(3)
outlier_table
```

```python
# Visualization of outliers
fig = px.histogram(outlier_table, x=['class_0_outlier_count', 'class_1_outlier_count'], y=outlier_table.index, barmode='stack')
fig.show()
```

I suspect the dataset may have multicollinearity issues, given the several high ($> |0.7|$) correlations between the features. However, it is expected, since the features are statistical measures derived from similar underlying data (integrated profile and DM-SNR curve). From statistics theory, those variables are not independent, hence the correlations.

In order to systematically identify multicollinearity, I will compute the Variance Inflation Factor (VIF) for each feature. VIF can be computed for each feature as:
$$\text{VIF} = \frac{1}{1 - R_i^2}.$$
Where \(R_i^2\) is the coefficient of determination of a regression of that feature against all the other features. A VIF value greater than 5 or 10 indicates a high multicollinearity that may be problematic.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Adds a constant column for VIF calculation
X = df.drop(columns=['target_class'])
X['constant'] = 1
```

```python
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
```

```python
# Removes the constant row
vif_data = vif_data[vif_data['feature'] != 'constant']
```

We have successfully computed the VIF for each feature and confirmed that multicollinearity is present in the dataset, as most features have a VIF greater than 10.

As a next step, we could consider removing some of the highly collinear features or applying dimensionality reduction techniques like Principal Component Analysis (PCA) to mitigate the multicollinearity issue. I'll go with PCA to condensate highly correlated variables in independent components without multicollinearity, while preserving most of the variance from the original data.
